```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('./data/hatecheck-data/test_suite_cases.csv',index_col=0)
```

```python
data['functionality'].value_counts()
```

```python
sns.countplot(data=data, x='functionality',order = data['functionality'].value_counts().index)
plt.xticks(rotation=90)
plt.xlabel('Functionality')
plt.ylabel('Frequency count')
plt.show()
```
